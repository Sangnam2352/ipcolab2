```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/MyDrive/ip/Images/shrinking.bmp')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gimg)
```

```python
outlmg = np.zeros(gimg.shape, dtype=np.uint8)
cv2_imshow(outlmg)
```

```python
scaling = 1.5

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    y = int(h * scaling)
    x = int(w * scaling)
    if y < gimg.shape[0] and x < gimg.shape[1]:
     outlmg[y,x] = gimg[h,w]

cv2_imshow(outlmg)
```

```python
blur = cv2.GaussianBlur(gimg, (5,5), 0)
cv2_imshow(blur)
outlmg = cv2.resize(gimg, None, fx=0.3, fy= 0.3, interpolation=cv2.INTER_AREA)
cv2_imshow(outlmg)
```

```python
img = cv2.imread('/content/gdrive/MyDrive/ip/Images/lena.jpg')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gimg)
```

```python
height, width = gimg.shape
print(height, width)
```

```python
M = cv2.getRotationMatrix2D((width/2.0, height/2.0), 45, 1)
print(M)
```

```python
outimg = cv2.warpAffine(gimg, M, (width, height))
cv2_imshow(outimg)
```

```python
M = np.float32([[1, 0, 30], [0, 1, 20]])
outimg = cv2.warpAffine(gimg, M, (width, height))
cv2_imshow(outimg)
```
