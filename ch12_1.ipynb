```python
img2 = img.copy()
```

```python
edges = cv2.Canny(glmg, 100, 200)
cv2_imshow(edges)
```

```python
lines = cv2.HoughLines(edges, 1, np.pi/180,220)

for line in lines:
  rho, theta = line[0]
  a = np.cos(theta)
  b = np.sin(theta)
  x0 = a * rho
  y0 = b * rho

  x1,y1 = int (x0 + 1000 * (-b)), int (y0 + 1000 * (a))
  x2, y2 = int (x0 - 1000 * (-b)), int (y0 - 1000 * (a))
  cv2.line(img2, (x1, y1), (x2, y2), (0, 0, 255), 2)

  cv2_imshow(img2)
```
