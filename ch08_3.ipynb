```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
lmg = cv2.imread('/content/gdrive/MyDrive/ip/Images/rice.bmp')
if lmg is None:
    print("Error: Image not found or could not be loaded. Please check the file path.")
else:
    print(lmg.shape)
    glmg = cv2.cvtColor(lmg, cv2.COLOR_RGB2GRAY)
    cv2_imshow(glmg)
```

```python
kernel = np.ones((3,3), np.uint8)
erosion = cv2.erode(glmg, kernel, iterations=1)
cv2_imshow(erosion)
```

```python
hist = cv2.calcHist([erosion], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
gBin = np.zeros((glmg.shape[0], glmg.shape[1]), dtype=np.uint8)

threshold = 120

for i in range(glmg.shape[0]):
  for j in range(glmg.shape[1]):
    if erosion[i, j] > threshold:
      gBin[i, j] = 255

cv2_imshow(gBin)
```

```python
num_labels, labels = cv2.connectedComponents(gBin, connectivity=8)
print(num_labels)
```

```python
kernel = np.ones((3,3), np.uint8)
dilate = cv2.dilate(erosion, kernel, iterations=1)
cv2_imshow(dilate)
```
