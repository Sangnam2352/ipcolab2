```python
img = cv2.imread('/content/gdrive/My Drive/images/lena256.jpg')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gimg)
```

```python
outimg = cv2.GaussianBlur(gimg, (7,7), 2)
cv2_imshow(outimg)
```

```python
edgeImage = cv2.addWeighted(gimg, 1, outimg, -1, 0)
cv2_imshow(edgeImage)
```

```python
unsharpImage = cv2.addWeighted(gimg, 1, edgeImage, 2, 0)
cv2_imshow(unsharpImage)
```

```python
kernel3 = np.ones((3,3), np.float32)/9
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = -1/9
kernel3[1][1] = 17/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
print(kernel3)
```
